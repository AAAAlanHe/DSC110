# data cleaning

```python
#where is the comment
df.isna().sum()
```

```python
df.isna().mean() *100
```

```python
# Drop rows where has_diabetes is missing
df_clean = df.dropna(subset=['has_diabetes']).copy()
```

```python
df_clean.isna().mean() *100
```

```python
df_clean.shape
```

```python
#Fill missing systolic_bp with mean
mean_bp = df_clean['systolic_bp'].mean()
print(mean_bp)
```

```python
df_clean['systolic_bp'] = df_clean['systolic_bp'].fillna(mean_bp)
```

```python
# Check it worked - should be 0 now
df_clean['systolic_bp'].isna().sum()
```

# Transform variables

```python
#converting numerical to categorical
# Create BMI categories
df['bmi_category'] = pd.cut(df['bmi'],
                             bins=[0, 18.5, 25, 30, 100],
                             labels=['Underweight', 'Normal', 'Overweight', 'Obese'])
```

```python
# Check it worked
df['bmi_category'].value_counts()
```

# Categorical Variables transformation

```python
df['education'].head()
```

```python
# Create a mapping dictionary (adjust based on your data dictionary)
education_map = {
    1: 'Less than 9th grade',
    2: '9-11th grade',
    3: 'High school grad',
    4: 'Some college or AA degree',
    5: 'College graduate',
    9: 'Do not know',
    None: 'Missing'
}
```

```python
education_map[1]
```

```python
# Apply the mapping
df['education_label'] = df['education'].map(education_map)
```

```python
# Check it worked
df['education_label'].value_counts()
```
